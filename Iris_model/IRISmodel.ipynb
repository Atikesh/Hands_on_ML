```python
#Now its time for handling Data 
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
df['Species']=encoder.fit_transform(df['Species'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns='Species')
X
```

```python
y=df['Species']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,
                                               random_state=123)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.tree import plot_tree
```

```python
from sklearn.metrics import accuracy_score , r2_score 
```

```python
dtc=DecisionTreeClassifier()
dtc2=DecisionTreeClassifier(criterion='entropy')
```

```python
dtc.fit(X_train,y_train)
dtc2.fit(X_train,y_train)
```

```python
plot_tree(dtc)
```

```python
plot_tree(dtc2)
```

```python
y_pred=dtc.predict(X_test)
y_pred2=dtc2.predict(X_test)
```

```python
print("Accuracy of Simple Decision Tree: ",accuracy_score(y_test,y_pred))
print("Accuracy of Decision Tree with Entropy criteria: ",
      accuracy_score(y_test,y_pred2))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [2, 3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(estimator=dtc, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best score:", grid_search.best_score_)
```

```python
grid_search2 = GridSearchCV(estimator=dtc2, param_grid=param_grid, cv=5)
grid_search2.fit(X_train, y_train)
print("Best score:", grid_search.best_score_)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create base classifier
base_classifier = DecisionTreeClassifier()

# Create BaggingClassifier
bagging_classifier = BaggingClassifier(estimator=base_classifier, n_estimators=10, random_state=42)

# Train the model
bagging_classifier.fit(X_train, y_train)

# Make predictions
predictions = bagging_classifier.predict(X_test)
accuracy_score(y_test,predictions)
```

```python
# Create base classifier
base_classifier2 = DecisionTreeClassifier()

# Create BaggingClassifier
bagging_classifier = BaggingClassifier(estimator=base_classifier,bootstrap=True, n_estimators=10, random_state=123)

# Train the model
bagging_classifier.fit(X_train, y_train)

# Make predictions
predictions = bagging_classifier.predict(X_test)
accuracy_score(y_test,predictions)
```

```python
# Create base classifier
base_classifier2 = DecisionTreeClassifier()

# Create BaggingClassifier
bagging_classifier = BaggingClassifier(estimator=base_classifier,bootstrap=True,oob_score=True, random_state=123)

# Train the model
bagging_classifier.fit(X_train, y_train)

# Make predictions
predictions = bagging_classifier.predict(X_test)
print("accuracy",accuracy_score(y_test,predictions))
print("oob_score",bagging_classifier.oob_score_)
```
