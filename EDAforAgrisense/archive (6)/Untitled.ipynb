```python
#mapping/encoding data categorical data
def map_soil_type(encoded_soil_type):
    soil_type_mapping = {
        0: 'sandy',
        1: 'loamy',
        2: 'black',
        3: 'red',
        4: 'clayey',
        5: 'dry',
        6: 'humid',
        7: 'wet'
    }
    return soil_type_mapping.get(encoded_soil_type, 'Unknown')

def map_fertilizer_name(encoded_fertilizer_name):
    fertilizer_name_mapping = {
        0: 'urea',
        1: 'dap',
        2: '14-35-14',
        3: '28-28',
        4: '17-17-17',
        5: '20-20',
        6: '10-26-26'
    }
    return fertilizer_name_mapping.get(encoded_fertilizer_name, 'Unknown')

def map_region(encoded_region):
    region_mapping = {
        0: 'desert',
        1: 'humid',
        2: 'semi arid',
        3: 'semi humid',
    }
    return region_mapping.get(encoded_region, 'Unknown')

def map_weather_condition(encoded_weather_condition):
    weather_condition_mapping = {
        0: 'normal',
        1: 'rainy',
        2: 'sunny',
        3: 'windy',
    }
    return weather_condition_mapping.get(encoded_weather_condition, 'Unknown')

# Apply mapping functions to the encoded columns
final_df['Soil Type'] = final_df['Soil Type'].apply(map_soil_type)
final_df['Fertilizer Name'] = final_df['Fertilizer Name'].apply(map_fertilizer_name)
final_df['Crop Name'] = final_df['Crop Name'].apply(map_crop_name)
final_df['REGION'] = final_df['REGION'].apply(map_region)
final_df['WEATHER CONDITION'] = final_df['WEATHER CONDITION'].apply(map_weather_condition)

# Rename columns for consistency
df_decoded = final_df.rename(columns={
    'Moisture': 'Moisture',
    'Humidity': 'Humidity',
    'N': 'N',
    'K': 'K',
    'P': 'P',
    'Temperature': 'Temperature',
    'ph': 'ph',
    'Rainfall': 'Rainfall',
    'CropDays': 'CropDays',
    'SoilMoisture': 'SoilMoisture',
    'Irrigation': 'Irrigation',
    'WATER REQUIREMENT': 'WATER REQUIREMENT'
})

# Reorder columns to match the original dataset
original_columns_order = ['Moisture', 'Soil Type', 'Fertilizer Name', 'Humidity', 'Crop Name', 'N', 'K', 'P', 'Temperature', 'ph', 'Rainfall', 'CropDays', 'SoilMoisture', 'Irrigation', 'REGION', 'WEATHER CONDITION', 'WATER REQUIREMENT']
df_decoded = df_decoded[original_columns_order]

# Display the decoded DataFrame
print(df_decoded)
```
